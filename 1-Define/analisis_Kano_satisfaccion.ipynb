```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 
```

```python
data = pd.read_excel('./Kano_Model.xlsx', header = 0)
# Mostrar las primeras filas para comprobar que se han cargado correctamente
print(data.head())
```

```python
# Supongamos que queremos analizar la frecuencia de una columna específica llamada 'columna_de_interes'
respuesta_column = 'Respuesta'
pregunta_column = 'Descripcion_pregunta'
numero_pregunta_column = 'Numero_pregunta'

# Calcular la frecuencia de los valores en la columna seleccionada
frecuencias_respuesta = data[respuesta_column].value_counts()
frecuencias_pregunta = data[pregunta_column].value_counts()

```

```python
plt.figure(figsize=(14, 8))  # Tamaño del gráfico
sns.barplot(x=frecuencias_pregunta.index, y=frecuencias_pregunta.values, label='Columna 1', alpha=0.7)
plt.title(f'Frecuencia de valores en la columna {respuesta_column} y {pregunta_column}')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)  # Rotar etiquetas en el eje x para mejor visualización
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 8))
ax = sns.countplot(data, x=pregunta_column, hue=respuesta_column, palette='Spectral')
plt.title('Frecuencia de respuestas por pregunta', fontweight='bold')
plt.xlabel('Descripción de pregunta', fontweight='bold')
plt.ylabel('Frecuencia', fontweight='bold')
plt.xticks(rotation=90)  # Rotar etiquetas en el eje x para mejor visualización
plt.legend(title='Respuesta', bbox_to_anchor=(1.05, 0), loc='lower left' )

# Configurar el intervalo en el eje y
ax.yaxis.set_major_locator(ticker.MultipleLocator(2))  # Mostrar valores de 2 en 2 en el eje y

# Agregar etiquetas a las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```
